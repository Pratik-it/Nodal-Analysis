```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#IPR Dataset (source: PetroWiki)
q_IPR= [0,1999,3094,3902,4512,4963,5275,5458,5519]
Pwf_IPR= [4000,3500,3000,2500,2000,1500,1000,500,14.7]
df_IPR= pd.DataFrame({'Pwf':Pwf_IPR, 'q':q_IPR})
df_IPR
```

```python
plt.plot(q_IPR,Pwf_IPR)
plt.xlabel('Q (bpd)')
plt.ylabel('Pwf (Psi)')
plt.title('IPR Curve')
```

```python
#TPR Dataset
q_TPR= np.arange(1000,6500,500)
P_190=[1334,1400,1487,1592,1712,1843,1984,2132,2287,2446,2689]
P_2375=[1298,1320,1351,1390,1435,1487,1545,1609,1677,1749,1824]
P_2875=[1286,1294,1305,1319,1336,1356,1378,1403,1431,1461,1493]
df_TPR= pd.DataFrame({'Q':q_TPR, 'P190':P_190, 'P2375':P_2375, 'P2875':P_2875})
df_TPR
```

```python
plt.style.use('default')
plt.figure(figsize=(10,5))
plt.plot(q_IPR,Pwf_IPR,label='IPR',linewidth='2')
plt.plot(q_TPR,P_190,label='TPR at 1.90in tubing',linewidth='2')
plt.plot(q_TPR,P_2375,label='TPR at 2.375in tubing',linewidth='2')
plt.plot(q_TPR,P_2875,label='TPR at 2.875in tubing',linewidth='2')
plt.xlabel('Q(MMscf/d)')
plt.ylabel('Flowing BHP(psi)')
plt.title('Nodal Analysis')
plt.legend(loc='best')
```
